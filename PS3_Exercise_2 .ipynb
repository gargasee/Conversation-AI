```python
!pip install wordcloud matplotlib
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
```

```python
# Load the text file
with open('JaneEyre.txt', 'r') as file:
    text = file.read()

# Customize stopwords by adding your own
custom_stopwords = set(STOPWORDS).union({"Mr", "Mrs", "Miss", "one", "said"})


# Create a WordCloud object with all customizations
wordcloud = WordCloud(
    width=800,
    height=400,
    background_color="black",  # Custom background color
    colormap="plasma",         # Custom colormap; try other options like 'viridis', 'magma'
    max_words=200,             # Limit the number of words displayed
    stopwords=custom_stopwords,  # Custom set of stopwords
    min_font_size=10
).generate(text)

# Display the customized word cloud
plt.figure(figsize=(12, 6))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```

```python
# Load the mask image
mask_path = '/content/test_image.jpeg'
mask = np.array(Image.open(mask_path))

# Create the word cloud with the mask
wordcloud = WordCloud(
    background_color='white',
    max_words=100,
    stopwords=STOPWORDS,
    mask=mask
).generate(text)

# Display the word cloud
plt.figure(figsize=(30, 10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()

```
